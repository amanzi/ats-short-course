```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
```

```python
# Create plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Sine wave')
plt.title('Matplotlib Test Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
print("Matplotlib is working correctly if you see the sine wave above.")
```
